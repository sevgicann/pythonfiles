```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatch
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({
    'Group':['A','B','C','D','E'],
    'Value':[1,5,4,3,9]
})
```

```python
plt.barh(y=df.Group,width=df.Value)
```

```python
df=df.sort_values(by=['Value'])
```

```python
plt.barh(y=df.Group,width=df.Value)
#plt.barh() işlevi, yatay çubuk grafiklerini çizmek için kullanılır. 
```

```python
plt.barh(
         y=df.Group,
         width=df.Value,
         color=['purple','skyblue'],
         edgecolor='black',
         linewidth=3)
```

```python

my_plot=plt.barh(
         y=df.Group,
         width=df.Value,
         color=mcolors.TABLEAU_COLORS,
         edgecolor='black',
         linewidth=3
         
         )
plt.bar_label(container=my_plot, padding=-10)

```

```python
#Benim yazdığım
tips = sns.load_dataset("tips")


sns.barplot(x="day", y="smoker", data=tips, color='blue', alpha=0.5, label='smoker')
sns.barplot(x="day", y="tip", data=tips, color='red', alpha=0.5, label='Tip')


plt.xlabel('Day')
plt.ylabel('Value')
plt.legend()
```

```python
tips.head()
```

```python
total=tips.groupby('day')['total_bill'].sum().reset_index()
```

```python
total
```

```python
bar1=sns.barplot(x="day",y="total_bill", data=total,color='darkblue')
```

```python
smoker=tips[tips.smoker=='Yes']
```

```python
bar1=sns.barplot(x="day",y="total_bill", data=total, color='darkblue')
bar2=sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)
top_bar=mpatch.Patch(color='darkblue',label='non-smoker')
bottom_bar=mpatch.Patch(color='lightblue',label='smoker=Yes')
plt.legend(handles=[top_bar,bottom_bar])
plt.show()
```

```python
df_iris=sns.load_dataset('iris')
```

```python
df_iris.head()
```

```python
sns.histplot(data=df_iris, x='sepal_length')
```

```python
sns.histplot(data=df_iris, x='sepal_length',kde=True)
```

```python
sns.displot(df_iris["sepal_length"])
#displot is deprecated
```

```python
sns.regplot(x=df_iris['sepal_length'],y=df_iris['sepal_width'])
```

```python
sns.regplot(
    x=df_iris['sepal_length'],
    y=df_iris['sepal_width'],
    marker="o", #Plot'un şeklini değiştirme
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5
    },
    scatter_kws={
        "color":"darkblue",
        "s":12
    }
    )
```

```python
sns.lmplot(
    x='sepal_length',
    y='sepal_width',
    data=df_iris,
    hue='species'
)
#Bu fonksiyon, veri noktalarının dağılımını gösterirken aynı zamanda 
# bu noktalar arasındaki ilişkiyi temsil etmek üzere bir regresyon çizgisi ekler.
```

```python
sns.boxplot(x=df_iris['species'],y=df_iris['sepal_length'])
#Kutu grafiği, veri setinin çeyrekliklerini, medyanını, minimum ve maksimum değerlerini ve aykırı değerleri gösteren bilgi
# dolu bir grafiktir. Bu grafiği kullanarak değişkenler arasındaki dağılımı karşılaştırabilir ve aykırı değerleri tespit edebilirsiniz.
```

```python
#Benim örneğim
import matplotlib.pyplot as plt
import numpy as np

# Örnek veri oluşturma
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Boxplot oluşturma
plt.boxplot(data)

# Grafik düzenlemeleri
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
