```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
```

```python
url="https://realpython.github.io/fake-jobs/"
page=requests.get(url) #Belirtilen url'a Get isteği gönderme
```

```python
#Karakter kodlamasıyla ilgili sorunları önlemek için .content eklendi
soup=BeautifulSoup(page.content,"html.parser")
```

```python
type(soup)
```

```python
# html_content içeriğini BeautifulSoup ile analiz eder ve girintili yapıda düzenler
print(soup.prettify())
```

```python
job_elements=soup.find_all("div",class_="card-content")
```

```python
for job_element in job_elements:
    title=job_element.find("h2",class_="title")
    location=job_element.find("p",class_="location")
    company=job_element.find("h3",class_="company")
    print(title)
    print(location)
    print(company)
```

```python
for job_element in job_elements:
    title=job_element.find("h2",class_="title")
    location=job_element.find("p",class_="location")
    company=job_element.find("h3",class_="company")
    print(title.text.strip()) #boşlukları temizlemek için kullanılır
    print(location.text.strip())
    print(company.text.strip())
    print("\n")
```

```python
python_jobs=soup.find_all("h2",string=lambda text:"python" in text.lower())
#string=lambda aranan kelimenin string ifadede geçiyorsa alınacağını gösterir
```

```python
for job in python_jobs:
    print(job.text)
```

```python
python_job_elements = [
    h2_element.parent.parent.parent for h2_element in python_jobs
]
```

```python
for job_element in job_elements:
    links = job_element.find_all('a')
    for link in links:
        if link.text == "Apply":
            link_url = link["href"]
            print(f"Apply here: {link_url}\n")
```

```python
job_list=[]
for job_element in job_elements:
    title=job_element.find("h2",class_="title")
    job_title=title.text.strip()
    print(job_title)
    job_list.append(job_title)
```

```python
len(job_list)
```

```python
df=pd.DataFrame(job_list,columns=["job_title"])
```

```python
location_list=[]
for job_element in job_elements:
    location_element=job_element.find("p",class_="location")
    location=location_element.text.strip()
    print(location)
    location_list.append(location)
```

```python
location=pd.Series(location_list, name="location")
df["location"]=location
df.sample(5)
```

```python
company_list=[]
for job_element in job_elements:
    company=job_element.find("h3",class_="company")
    company=company.text.strip()
    print(company)
  
    company_list.append(company)

  
```

```python
company=pd.Series(company_list, name="company")
df["company"]=company
df.sample(5)
df.head()
```
