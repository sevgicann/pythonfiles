```python
url = 'https://xeno-canto.org/api/2/recordings?query=troglodytes+troglodytes'
```

```python
#requests get methodunu kullanarak ilgili url'den veri alıyoruz
try:
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()  # API'den gelen veriyi JSON olarak alma

        if 'recordings' in data:
            # Verilerde almak istediğim kısımlar için bir döngü oluşturuyorum
            for recording in data['recordings']:
                print(f"Kayıt ID: {recording['id']}")
                print(f"Kuş Türü: {recording['en']}")
                print(f"Ülke: {recording['cnt']}")
                print(f"Konum: {recording['loc']}")
                print(f"Kayıt URL'si: {recording['file']}")
                print("\n\n")
        else:
            print("Kayıt bulunamadı.")
    else:
        print(f"İstek durumu: {response.status_code}")
        print(f"Hata mesajı: {response.text}")  # Hata mesajı alma 
except requests.RequestException as e:
    print(f"Hata: {e}")
```

```python

def get_recordings_data(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()  # API'den gelen veriyi JSON olarak alma

            if 'recordings' in data:
                # Verilerde almak istediğim kısımlar için bir döngü oluşturuyorum
                for recording in data['recordings']:
                    yield {
                        'Kayıt ID': recording['id'],
                        'Kuş Türü': recording['en'],
                        'Ülke': recording['cnt'],
                        'Konum': recording['loc'],
                        'Kayıt URL\'si': recording['file']
                    }
            else:
                yield "Kayıt bulunamadı."
        else:
            yield f"İstek durumu: {response.status_code}"
            yield f"Hata mesajı: {response.text}"  # Hata mesajı alma 
    except requests.RequestException as e:
        yield f"Hata: {e}"

```

```python

def get_recordings_data(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()  

            if 'recordings' in data:
                recordings_list = []
                
                for recording in data['recordings']:
                    recording_data = {
                        'Kayıt ID': recording['id'],
                        'Kuş Türü': recording['en'],
                        'Ülke': recording['cnt'],
                        'Konum': recording['loc'],
                        'Kayıt URL\'si': recording['file']
                    }
                    recordings_list.append(recording_data)

                return pd.DataFrame(recordings_list)
            else:
                return pd.DataFrame(["Kayıt bulunamadı."])
        else:
            return pd.DataFrame([f"İstek durumu: {response.status_code}", f"Hata mesajı: {response.text}"])
    except requests.RequestException as e:
        return pd.DataFrame([f"Hata: {e}"])



```
