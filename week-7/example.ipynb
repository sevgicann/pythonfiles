```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 3 + 2*(x)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
```

Exponential

```python
X = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

Y= np.exp(X)
```

```python
plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

Logarithmic
y = log(x)

Instead of x, we can use X, which can be a polynomial representation of x values.
y = log(X)

```python
X = np.arange(1.0, 100.0, 0.1)

Y = np.log(X)
```

```python
plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

Logistic / Sigmoid
Y = a + (b / 1 + c ^ (X-d))

```python
X = np.arange(-5.0, 5.0, 0.1)

Y = 1-4/(1+np.power(3, X-2))
```

```python
plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Veriyi ykleme
df = pd.read_csv("china_gdp.csv")

# Veriyi inceleme
print(df.head())

# Veriyi yeniden ekillendirme
X = df["Year"].values
y = df["Value"].values  # "Value" stunu ile gncellendi

# ssel fonksiyon tanmlama (y = a * exp(b * x))
def exponential_func(x, a, b):
    return a * np.exp(b * x)

# Modeli veriye uydurma
params, covariance = curve_fit(exponential_func, X, y)

# Modelin parametreleri
a, b = params

# Grafik oluturma
plt.figure(figsize=(8, 6))
plt.scatter(X, y, label="Veri")
plt.plot(X, exponential_func(X, a, b), color='red', label="ssel Model")
plt.title("in GDP Verisi ve ssel Model")
plt.xlabel()
```
