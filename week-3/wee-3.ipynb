```python
import matplotlib.pyplot as plt
import numpy as np

# x değerlerini oluştur
x = np.linspace(0, 2*np.pi, 100)  # 0 ile 2*pi arasında 100 adet x değeri oluştur

# sinüs fonksiyonu kullanarak y değerlerini hesapla
y = np.sin(x)

# Grafiği çiz
plt.figure(figsize=(8, 6))  # Grafik boyutunu belirle
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)  # x'e karşılık gelen sinüs değerlerini çiz

plt.title('Sinüs Grafiği')  # Grafik başlığı
plt.xlabel('x')  # x ekseninin etiketi
plt.ylabel('sin(x)')  # y ekseninin etiketi
plt.legend()  # Gösterilen çizginin adını göster

# Grafiği göster
plt.grid(True)  # Izgara çizgilerini göster
plt.show()
```
