```python

df_recovered=df_recovered.drop(columns=['Lat', 'Long', 'Province/State'])
df_deaths=df_deaths.drop(columns=['Lat', 'Long', 'Province/State'])
# Verilen değerleri geçici olarak siler. inpalce=false olduğu için geçici
```

```python
def make_frame(Country:str):
    """Take country name as parameter, return infected, recovered, deaths values as timeseries table.
    
    """

    df = pd.DataFrame({
        'infected' : df_infected.loc[Country],
        'recovered': df_recovered.loc[Country],
        'deaths': df_deaths.loc[Country]
    })
    df.index = pd.to_datetime(df.index, format='%m/%d/%y')
    return df
#Belirtilen ülke için yeni dataframe oluşturur
```

```python
df_us = make_frame('US')
df_turkey = make_frame('Turkey')
```

```python
df_us['new_infected'] = df_us['infected'].diff()
#diff() fonksiyonu, bir Pandas Serisi veya DataFrame'deki ardışık değerler arasındaki farkları hesaplamak için kullanılan bir metottur.
```

```python
df_us['new_infected_avg'] = df_us['new_infected'].rolling(window=7).mean()
```

```python
countries[(countries['Country_Region'] == 'US') & (countries['Province_State'].isna())]['Population']
#Bu ifade, 'countries' adlı DataFrame içinde 'Country_Region' sütunu 'US' olan ve 'Province_State' sütunu eksik (NaN) olan satırları filtreler ve bu satırlara ait 'Population' (Nüfus) sütununu seçer.
```

```python
pop_us = countries[(countries['Country_Region'] == 'US') & (countries['Province_State'].isna())]['Population'].iloc[0]
#Bu ifade, 'countries' adlı DataFrame içinde 'Country_Region' sütunu 'US' olan ve 'Province_State' sütunu eksik (NaN) olan ilk satırın 'Population' (Nüfus) sütununu seçer ve bu değeri 'pop_us' adlı değişkene atar.
```

```python
print(pop_us)
```

```python
df_us.head()
```

```python
df_us['pop_infected'] = df_us['infected'] * 100 / pop_us
#Bu ifade, 'df_us' DataFrame'inde bulunan 'infected' (enfekte olan kişi sayısı) sütunundaki değerleri, 'pop_us' adlı bir 
# önceki hesaplanmış nüfus değerine bölerek, her bir zaman dilimindeki enfekte olan kişi sayısının popülasyon içindeki yüzdesini hesaplar.
```

```python
df_us
```

```python
df_us['pop_infected'].plot()
plt.show()
```

```python
# The basic reproduction number
df_us['Rt'] = df_us['new_infected'].rolling(8).apply(lambda x: x[4:].sum() / x[:4].sum())
#Bu ifade, 'df_us' DataFrame'inde bulunan 'new_infected' (yeni enfekte olan kişi sayısı) sütunundaki veriler üzerinde 'Rt' (reprodüksiyon sayısı) 
# hesaplamak için bir işlem gerçekleştirir.
```

```python
df_us['Rt'].plot()
```

```python
df_us['Rt'].replace(np.inf, np.nan).fillna(method='bfill')
#Bu ifade, 'df_us' DataFrame'inde bulunan 'Rt' (Reprodüksiyon Sayısı) sütunundaki değerlerdeki
#  sonsuz (inf) değerleri NaN ile değiştirir ve ardından bu eksik değerleri bir sonraki değerle doldurur (backfill yöntemiyle).
```
